Hill and Valley Prediction with Logistic Regression

*   To predict whether it is Hill or Valley
DataSource:https://github.com/YBIFoundation/Dataset/blob/main/Hill%20Valley%20Dataset.csv



Import Library

```python
import pandas as pd
```

Import data

```python
df=pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/Hill%20Valley%20Dataset.csv")
```

Describe data

```python
df.describe()
```

Data Visualization

```python
df.head()
```

 Target Variable (y) and Feature Variables (X)

```python
x=df.drop(['Class'], axis=1)
y=df['Class']
y.value_counts()
```

Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.07,random_state=2526)
```

Modeling


```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

Model Evaluation

```python
model.fit(x_train,y_train)
```

Prediction

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
